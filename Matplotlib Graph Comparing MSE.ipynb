```python
import matplotlib.pyplot as plt
import numpy as np

models = ["Linear Regression", "Decision Tree", "ANN"]
mse_values = [10, 15, 8, 12]

mse_values = {'Training': (697, 187, 1098),
             'Testing': (664, 114, 1115)}

x = np.arange(len(models))


plt.bar(x -0.2, mse_values['Training'],  width = 0.2, color = 'cyan')
plt.bar(x, mse_values['Testing'], width = 0.2)
plt.xticks(x, models)
plt.legend(mse_values)
plt.xlabel("Models")
plt.ylabel("MSE")
plt.title("Mean Squared Error Comparison By Model")


plt.show()
```
